```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import xlabel

df = pd.read_csv('../data/Data.csv')

df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
((df.isnull().sum() /len(df)) *100).sort_values(ascending=False)
```

```python
per1 = df[df[['Количество призёров', 'Количество участников', 'Процент призёров']].isnull().sum(axis=1) >= 2]
per1
```

```python
df.drop(per1.index, inplace=True)
```

```python
df
```

```python
df = df[(df['Количество участников'].notna()) & (df['Количество участников'] != 0) & 
                 (df['Процент призёров'].notna()) & (df['Процент призёров'] != 0)]

df
```

```python
df_error = df[df.apply(lambda row: row.astype(str).str.contains('ошибка').any(), axis=1)]
df_error
```

```python
df = df[~df.apply(lambda row: row.astype(str).str.contains('ошибка').any(), axis=1)]
df
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.dropna(subset=['Средний балл по этапу'])
df
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['Количество участников'] = df['Количество участников'].astype(int)
df['Количество призёров'] = df['Количество призёров'].astype(int)
df['Процент призёров'] = df['Процент призёров'].astype(float)
df['Средний балл по этапу'] = df['Средний балл по этапу'].astype(float)
df['Год проведения'] = df['Год проведения'].astype(int)
```

```python
columns = 'Средний балл по этапу'

plt.figure(figsize=(10, 5))
sns.histplot(df[columns], bins=60, kde=True, color='darkblue')
plt.title(f'Гистограмма для {columns}')
plt.xlabel('Количество')
plt.ylabel('Частота')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x="Этап олимпиады", y="Средний балл по этапу", data=df)
plt.title("Распределение среднего балла по этапам олимпиады")
plt.xticks(rotation=45)  # Поворачиваем подписи оси X для удобства
plt.show()
```
